## 1. Create a 4 dimensional data set with 64 elements and show all 4 scatter 2D plots of the data $x_1$ vs. $y$, $x_2$ vs. $y$,  $x_3$ vs. $y$, $x_4$ vs. $y$

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
n = 64 
x = np.linspace(0, 1, n) + np.random.rand(4, n)
x = np.vstack([x, np.ones(len(x.T))]).T + 1
y = np.linspace(0, 1, n) + np.random.rand(n) - 1.7
```

```python
fig1, ax1 = plt.subplots()
ax1.scatter(x.T[0], y)
ax1.set(xlabel='x1', ylabel='y', title='x1 vs. y')
```

```python
fig2, ax2 = plt.subplots()
ax2.scatter(x.T[1], y)
ax2.set(xlabel='x2', ylabel='y', title='x2 vs. y')
```

```python
fig3, ax3 = plt.subplots()
ax3.scatter(x.T[2], y)
ax3.set(xlabel='x3', ylabel='y', title='x3 vs. y')
```

```python
fig4, ax4 = plt.subplots()
ax4.scatter(x.T[3], y)
ax4.set(xlabel='x3', ylabel='y', title='x4 vs. y')
```

## 2.  Create a Linear Regression model (like we did in class) to fit the data. *Use the example from Lesson 3 and do not use a library that calculates automatically*. We are expecting 5 coefficients to describe the linear model. 

## After creating the model (finding the coefficients), create a new column $y_p=\Sigma\beta_n\cdot x_n$ 

```python
left = np.linalg.inv(np.dot(x.T, x))
right = np.dot(y.T, x)
beta = np.dot(left, right)

yp = np.dot(x, beta)
yp
```

## 3. Plot the model's prediction as a different color on top of the scatter plot from Q1 in 2D for all 4 of the dimensions ($x_1 \rightarrow y_p, x_2 \rightarrow y_p, x_3 \rightarrow y_p, x_4 \rightarrow y_p$) 

```python
fig1, ax1 = plt.subplots()
ax1.scatter(x.T[0], y, c ='blue', label = 'Orginal')
ax1.scatter(x.T[0], yp, c='red', label = 'Pred')
ax1.set(xlabel='x1', ylabel='yp', title='x1 --> yp')
ax1.legend()
```

```python
fig2, ax2 = plt.subplots()
ax2.scatter(x.T[1], y, c ='blue', label = 'Orginal')
ax2.scatter(x.T[1], yp, c='red', label = 'Pred')
ax2.set(xlabel='x2', ylabel='yp', title='x2 --> yp')
ax2.legend()
```

```python
fig3, ax3 = plt.subplots()
ax3.scatter(x.T[2], y, c ='blue', label = 'Orginal')
ax3.scatter(x.T[2], yp, c='red', label = 'Pred')
ax3.set(xlabel='x3', ylabel='yp', title='x3 --> yp')
ax3.legend()
```

```python
fig4, ax4 = plt.subplots()
ax4.scatter(x.T[3], y, c ='blue', label = 'Orginal')
ax4.scatter(x.T[3], yp, c='red', label = 'Pred')
ax4.set(xlabel='x4', ylabel='yp', title='x4 --> yp')
ax4.legend()
```

## 4. Read in `mlnn/data/Credit.csv` with Pandas and build a Linear Regression model to predict Credit Rating (`Rating`). Use only the numeric columns in your model, but feel free to experiment which which columns you believe are better predicters of Credit Rating (Column `Rating`)

```python
import pandas as pd
import numpy as np
credit = pd.read_csv('../data/Credit.csv')
credit.head()
```

## Choose multiple columns as inputs beyond `Income` and `Limit` but clearly, don't use `Rating`

```python
columns = ['Income', 'Limit', 'Age', 'Education']
X = credit[columns].values

X = np.vstack([X.T, np.ones(len(X))]).T
X
```

```python
y = credit['Rating']
y
```

```python
left = np.linalg.inv(np.dot(X.T, X))
right = np.dot(y.T, X)
beta = np.dot(left, right)

income_pred = np.dot(X, beta)
income_pred
```

### 5. Plot your results using scatter plots (just like in class). Show as many of your columns vs. credit rating that you can.

```python
fig_in, ax_in = plt.subplots()
ax_in.scatter(X.T[0], y, label = 'Orginal')
ax_in.scatter(X.T[0], income_pred, label = 'Pred')
ax_in.set(xlabel='Income', ylabel='Rating', title='Income --> Rating')
ax_in.legend()
```
